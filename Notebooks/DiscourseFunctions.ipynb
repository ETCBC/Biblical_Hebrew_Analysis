It is, therefore, of the utmost importance to correctly identify the boundaries of communicative and linguistic domains on the basis of specific parameters and linguistic signals, such as the presence of conjunctions, <a href="MDModifier.ipynb" target="_blank">multiple-duty modifiers</a>, and developments in the set of <a href="Participants.ipynb" target="_blank">participants</a>. For a more elaborate discussion of markers of continuity and discontinuity in communicative and linguistic domains, we refer the reader to section 6.2 of the dissertation.</p>
