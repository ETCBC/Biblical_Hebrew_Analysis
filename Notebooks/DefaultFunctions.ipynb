# Default Functions of Clause Types in terms of Volitivity and Non-volitivity

<p>A central assumption in our theory on the verbal system of Biblical Hebrew is that the function of expressing volitive meaning is syntactically marked in <em>yiqtol</em> clauses. More specifically, we hold that it is the position of the verbal <em>yiqtol</em> form in these clauses which is a decisive criterion for the distribution of volitive and non-volitive functionality among <em>yiqtol</em> clauses.</p>
<p>If the <em>yiqtol</em> form takes clause-initial position - i.e.: if the verbal form is the first element in the clause (apart from the clause-opening coordinate conjunction <span class="hebrew">&#1493;</span>) - the clause has the expression of volitivity as its default functionality. This is true for all clauses of the types <strong>ZYqX</strong>, <strong>ZYq0</strong>, <strong>WYqX</strong>, and <strong>WYq0</strong>.</p>
<p>In our analyses we identify several categories of volitivity, which are related to the person of the subject (for our (automatically generated) renderings of the volitive forms, see the <a href="Translation.ipynb" target="_blank">'Translation' notebook</a>):</p>
<table class="presentation">
<tr><td>1st person</td><td>cohortative (<strong>coh.</strong>)</td><td>"Let me/us do..."</td></tr>
<tr><td>2nd person</td><td>jussive (<strong>juss.</strong>)</td><td>"You should do..."</td></tr>
<tr><td>3rd person</td><td>volitive (<strong>volit.</strong>)</td><td>"May he/she/it/they do..." / "Let him/her/it/them do..."</td></tr>
</table>

<p>If, on the other hand, the verbal <em>yiqtol</em> form is preceded by a constituent or another element than the coordinate conjunction <span class="hebrew">&#1493;</span>, its clause has the communication of non-volitive (<strong>non-vol.</strong>) meaning as its default function. This is true for all other types of <em>yiqtol</em> clauses: <strong>XYqt</strong>, <strong>XxYq</strong>, <strong>xXYq</strong>, <strong>WXYq</strong>, <strong>WXxY</strong>, <strong>WxXY</strong>, <strong>xYqX</strong>, <strong>xYq0</strong>, <strong>WxYX</strong>, and <strong>WxY0</strong>.</p> 

<p>For a correct assignment of default functions one has to take notice of <a href="CCR.ipynb" target="_blank">clause constituent relations</a> and specific relations between clause atoms. Thus, a verb-initial <em>yiqtol</em> clause which is preceded by a daughter subject clause, should be treated as an <strong>X-yiqtol</strong> clause (or sentence; cf. our <a href="ClauseLabels.ipynb" target="_blank">introduction to clausetype labels</a>) and should therefore be assigned a non-volitive meaning. We provide Ps 2.4 as an illustration:</p>

<table class="presentation">
<tr><td>004</td><td class="hebrew" nowrap>[&lt;Lo&gt; ב--שׁמימ]  [&lt;PC&gt; יושׁב]&nbsp;&nbsp;&nbsp;</td><td style="font-family:Courier">Ptcp</td><td nowrap>&nbsp;&nbsp;&nbsp;he/she who is sitting (in)</td></tr>
<tr><td></td><td class="hebrew" nowrap>[&lt;Pr&gt; ישׂחק]</td><td style="font-family:Courier"><strong>XYqt</strong></td><td nowrap>laughs</td></tr>             
</table>

<p>The default function labels <strong>coh.</strong>, <strong>juss.</strong>, <strong>volit.</strong>, and <strong>non-vol.</strong> only apply to <em>yiqtol</em> clauses and prospective <em>weqatal</em> clauses. For the default and final functions of other clause types we have used other labels:</p>
<table class="presentation">
<tr><td>imperative (<strong>imperat.</strong>)</td><td><em>imperative</em> clauses</td></tr>
<tr><td>infinitive (<strong>inf.</strong>)</td><td><em>infinitive</em> clauses</td></tr>
<tr><td>indicative (<strong>ind.</strong>)&#42;</td><td><em>qatal</em> clauses, <em>participle</em> clauses</td></tr>
<tr><td>nominal (<strong>nomin.</strong>)</td><td><em>nominal</em> and <em>adjective</em> clauses</td></tr>
<tr><td>existence (<strong>exist.</strong>)</td><td><em>nominal</em> clauses containing the copula <span class="hebrew">&#1497;&#64298;</span></td></tr>
<tr><td>non-existence (<strong>ng.exst.</strong>)</td><td><em>nominal</em> clauses containing either of the negative copulas <span class="hebrew">&#1488;&#1497;&#1503;</span> and <span class="hebrew">&#1489;&#1500;&#1497;</span></td></tr>
<tr><td>prohibitive existence (<strong>prh.exst</strong>)</td><td><em>nominal</em> clauses containing the volitive negation <span class="hebrew">&#1488;&#1500;</span></td></tr>
</table>

<p>In addition, the volitive or non-volitive functions of <em>yiqtol</em> and <em>weqatal</em> clauses are sometimes further specified by the following labels:</p>
<table class="presentation">
<tr><td>final&#42;&#42; (<strong>final</strong>)</td><td><em>weyiqtol</em> clauses</td></tr>
<tr><td>consecutive&#42;&#42; (<strong>consec.</strong>)</td><td><em>weqatal</em> clauses</td></tr>
<tr><td>prohibitivity (<strong>prohib.</strong>)</td><td>clauses containing (or inheriting) the volitive negation <span class="hebrew">&#1488;&#1500;</span></td></tr>
<tr><td>negative finality (<strong>neg.fin.</strong>)</td><td>clauses containing (or inheriting) the subordinate conjunction <span class="hebrew">&#1508;&#1503;</span></td></tr>
</table>

<p>Defective clause atoms, elliptic clauses, vocative clauses, casus pendentes, extrapositions, and macro-syntactic signals are assigned neither default functions nor final functions.</p>

<p class="note">&#42;We do not assume a functional difference between the labels <strong>'indicative'</strong> and <strong>'non-volitive'</strong>, but have decided to use the label <strong>'non-volitive'</strong> (instead of <strong>'indicative'</strong>) only for those clause types whose functionality may be influenced by <a href="Processes.ipynb" target="_blank">processes of inheritance or blocking</a>. In this way, discrepancies between default and final functions of clauses caused by such mechanisms can be identified most easily.</p>
