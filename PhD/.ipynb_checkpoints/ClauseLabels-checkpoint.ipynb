# Clause Type Labels

<p>The type of a clause depends, in first instance, on the presence of a verbal form in the clause. If such a verbal form is absent,
but the clause <em>does</em> contain a predicate, the clause is of a nominal kind and can be either a simple nominal clause (if the predicate 
is a noun phrase) - <strong>NmCl</strong> - or an adjective clause (if the predicate is an adjective phrase) - <strong>AjCl</strong>.</p>

<p>We identify the following types of clauses without a predicate:</p>
<table class="presentation">
<tr><td>Vocative clause</td><td><strong>Voct</strong></td></tr>
<tr><td>Casus Pendens</td><td><strong>CPen</strong></td></tr>
<tr><td>Ellipsis</td><td><strong>Ellp</strong></td></tr>
<tr><td>Macrosyntactic signal</td><td><strong>MSyn</strong></td></tr>
<tr><td>Extraposition</td><td><strong>XPos</strong></td></tr>
</table>

<p>The clause type of verbal clauses (i.e.: clauses with a verbal predicate) is determined by four factors:</p>
<ul>
<li>the presence or absence of a coordinate conjunction (<span class="hebrew">&#1493;</span>)</li>
<li>the presence or absence of constituents that precede the predicate <em>(indicated by capital <strong>X</strong> for subjects, small <strong>x</strong> for non-subject constituents, and <strong>Z</strong> for no (zero) preverbal constituents)</em></li>
<li>the verbal tense of the predicate, which can be either imperative, imperfect (<em>yiqtol</em>), perfect (<em>qatal</em>), participle, or infinitive</li>
<li>the presence or absence of an explicit subject following the predicate (indicated by capital <strong>X</strong> resp. <strong>0</strong>)</li>
</ul>

<p>In case of verbal clauses containing a non-finite verbal form, only the third factor (i.e.: the verbal tense of the predicate), is decisive for
the clause type of the clause. Thus, the following non-finite clause types can be identified:</p>
<table class="presentation">
<tr><td>Participle clause (both participle active and participle passive)</td><td><strong>Ptcp</strong></td></tr>
<tr><td>Infinitive construct clause</td><td><strong>InfC</strong></td></tr>
<tr><td>Infinitive absolute clause</td><td><strong>InfA</strong></td></tr>
</table>

<p>The clause type labels do not always reflect the <em>clause atom type</em> of a clause atom. There are two types of exceptional situations in which the clause type label takes into account the broader context of the clause or the sentence.</p>
<ol>
<li>If a clause is interrupted by an embedded clause and therefore consists of a defective (predicate-less) and a non-defective (with predicate) part, then the type of the clause as a whole is presented in the clause type label provided for the non-defective part. An example can be found in Ps 1.5:
<table class="presentation">
<tr><td>005</td><td class="hebrew" nowrap>[&lt;Su&gt; כל]  [&lt;Cj&gt; -ו]</td><td style="font-family:Courier">Defc</td><td nowrap>and/but all</td></tr>
<tr><td></td><td class="hebrew" nowrap>[&lt;Pr&gt; יעשׂה]  [&lt;Re&gt; אשׁר]&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;</td><td style="font-family:Courier">xYq0</td><td nowrap>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;for whom/which is true that he makes/does</td></tr>             
<tr><td></td><td class="hebrew" nowrap>[&lt;Pr&gt; יצליח]&nbsp;&nbsp;&nbsp;</td><td style="font-family:Courier">WXYq</td><td nowrap>&nbsp;&nbsp;&nbsp;is strong/succesful</td></tr>
</table>
Note that the conjunction and the explicit subject present in the defective clause atom are accounted for in the clause type label (<strong>WXYq</strong>) of the verb-initial non-defective part of the clause.</li>
<li>If a clause is preceded by a constituent daughter clause, its clause type label functions as a <em>'sentence type label'</em>, in which the constituent clause is treated as a simple constituent preceding the main clause's predication. As an illustration, we refer to Ps 2.4, where the subject participle daughter clause is represented in the clause type label of its subsequent <em>yiqtol</em> mother clause by the capital <strong>X</strong>:
<table class="presentation">
<tr><td>004</td><td class="hebrew" nowrap>[&lt;Lo&gt; ב--שׁמימ]  [&lt;PC&gt; יושׁב]&nbsp;&nbsp;&nbsp;</td><td style="font-family:Courier">Ptcp</td><td nowrap>&nbsp;&nbsp;&nbsp;he/she who is sitting (in)</td></tr>
<tr><td></td><td class="hebrew" nowrap>[&lt;Pr&gt; ישׂחק]</td><td style="font-family:Courier">XYqt</td><td nowrap>laughs</td></tr>             </table>
For more information on clause constituent relations, see <a href="CCR.ipynb" target="_blank">here</a>.
</li>
</ol>
